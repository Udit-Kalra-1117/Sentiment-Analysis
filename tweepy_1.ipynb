```python
from textblob import TextBlob
from matplotlib import pyplot as plt
import numpy as np
import re
```

```python
f = open("addedtweets.txt", "r", encoding="utf-8")
Lines = f.readlines()

sentili=[]

for i in Lines:
    sentili.append(TextBlob(i).sentiment.polarity)
    
print(sentili)

neg=0
n=0
p=0
c=0

for i in sentili:
    c=c+1
    if i==0.0:
        n=n+1
    elif i>0.0:
        p=p+1
    else:
        neg=neg+1
        
n1=(n/c)*100
p1=(p/c)*100
neg1=(neg/c)*100

print(n1,p1,neg1)

y = np.array([n1, p1, neg1])
mylabels = ["Neutral", "Positive", "Negative"]

plt.pie(y, labels = mylabels, autopct='%1.1f%%',
        shadow=True,explode=[0.1,0.2,0.1])
plt.show()
    
```
